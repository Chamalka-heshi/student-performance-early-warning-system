```python
import pandas as pd

df = pd.read_csv("../data/processed/students_cleaned.csv")
df.shape
```

```python
df_features = df.drop(columns=['GPA'])
df_features.columns.tolist()
```

```python
X = df_features.drop(columns=['GradeClass'])
y = df_features['GradeClass']

print(X.shape, y.shape)
X.columns.tolist()
```

```python
X = pd.get_dummies(X, columns=['Ethnicity'], prefix='Ethnicity')
X.columns.tolist()
```

```python
X['HighAbsenceFlag'] = (df_features['Absences'] > 15).astype(int)
X[['Absences', 'HighAbsenceFlag']].head()
```

```python
print("Final features:", X.columns.tolist())
print("Shape:", X.shape)
X.head()
```

```python
import os
os.makedirs("../data/processed", exist_ok=True)
X.to_csv("../data/processed/X_features.csv", index=False)
y.to_csv("../data/processed/y_target.csv", index=False)
```
