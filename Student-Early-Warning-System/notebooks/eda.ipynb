```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv("../data/processed/students_cleaned.csv")
df.head()
```

```python
plt.figure(figsize=(7,5))
sns.countplot(x='GradeClass', data=df, palette='viridis')
plt.title('Distribution of GradeClass')
plt.xlabel('Grade Class (0=Best, 4=Worst)')
plt.ylabel('Number of Students')
plt.savefig('../images/gradeclass_distribution.png', dpi=150, bbox_inches='tight')
plt.show()
```

```python
plt.figure(figsize=(8,5))
sns.boxplot(x='GradeClass', y='Absences', data=df, palette='viridis')
plt.title('Absences by Grade Class')
plt.xlabel('Grade Class (0=Best, 4=Worst)')
plt.ylabel('Number of Absences')
plt.savefig('../images/absences_by_gradeclass.png', dpi=150, bbox_inches='tight')
plt.show()
```

```python
plt.figure(figsize=(8,5))
sns.boxplot(x='GradeClass', y='StudyTimeWeekly', data=df, palette='viridis')
plt.title('Weekly Study Time by Grade Class')
plt.xlabel('Grade Class (0=Best, 4=Worst)')
plt.ylabel('Study Time (hours/week)')
plt.savefig('../images/studytime_by_gradeclass.png', dpi=150, bbox_inches='tight')
plt.show()
```

```python
plt.figure(figsize=(12,9))
corr = df.corr(numeric_only=True)
sns.heatmap(corr, annot=True, fmt='.2f', cmap='coolwarm', center=0)
plt.title('Correlation Heatmap')
plt.savefig('../images/correlation_heatmap.png', dpi=150, bbox_inches='tight')
plt.show()
```

```python
plt.figure(figsize=(8,5))
sns.boxplot(x='GradeClass', y='ParentalSupport', data=df, palette='viridis')
plt.title('Parental Support by Grade Class')
plt.savefig('../images/parentalsupport_by_gradeclass.png', dpi=150, bbox_inches='tight')
plt.show()
```

```python
plt.figure(figsize=(8,5))
sns.countplot(x='Tutoring', hue='GradeClass', data=df, palette='viridis')
plt.title('Tutoring vs Grade Class')
plt.savefig('../images/tutoring_by_gradeclass.png', dpi=150, bbox_inches='tight')
plt.show()
```
